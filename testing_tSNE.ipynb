## Testing Out t-SNE

```python
import pandas as pd 
import plotly.express as px
from sklearn.datasets import make_classification

X, y = make_classification(
    n_features=6,
    n_classes=3,
    n_samples=1500,
    n_informative=2,
    random_state=5,
    n_clusters_per_class=1,
)

fig = px.scatter_3d(x=X[:, 0], y=X[:, 1], z=X[:, 2], color=y, opacity=0.8)
fig.show()
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, random_state=42)
X_tsne = tsne.fit_transform(X)
tsne.kl_divergence_
```

```python
fig = px.scatter(x=X_tsne[:, 0], y=X_tsne[:, 1], color=y)
fig.update_layout(
    title="out-of-the-box t-SNE visualization of Custom Classification dataset",
    xaxis_title="First t-SNE",
    yaxis_title="Second t-SNE",
)
fig.show()
```

```python
# number of points 
n = 1500
tsne_250 = TSNE(n_components=2, early_exaggeration=n/10, learning_rate=1, max_iter=250,random_state=42)
X_tsne_250 = tsne_250.fit_transform(X)
tsne_250.kl_divergence_
```

```python
tsne_500 = TSNE(n_components=2, early_exaggeration=n/10, learning_rate=1, max_iter=5000,random_state=42)
X_tsne_500 = tsne_500.fit_transform(X)
tsne_500.kl_divergence_
```

```python
fig = px.scatter(x=X_tsne_250[:, 0], y=X_tsne_250[:, 1], color=y)
fig.update_layout(
    title="t-SNE visualization with improved alpha and h (250 iterations)",
    xaxis_title="First t-SNE",
    yaxis_title="Second t-SNE",
)
fig.show()
```

```python
fig = px.scatter(x=X_tsne_500[:, 0], y=X_tsne_500[:, 1], color=y)
fig.update_layout(
    title="t-SNE visualization with improved alpha and h (5000 iterations)",
    xaxis_title="First t-SNE",
    yaxis_title="Second t-SNE",
)
fig.show()
```
