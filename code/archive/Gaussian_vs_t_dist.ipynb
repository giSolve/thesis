```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, t

# Define the x range
x = np.linspace(-5, 5, 1000)

# Compute the probability density functions (PDFs)
gaussian_pdf = norm.pdf(x, loc=0, scale=1)  # Standard normal (mean=0, std=1)
cauchy_pdf = t.pdf(x, df=1)  # Student's t with df=1 (Cauchy distribution)

# Plot the distributions
plt.figure(figsize=(10, 5))
plt.plot(x, gaussian_pdf, label="Normal Distribution", color="blue", linewidth=2)
plt.plot(x, cauchy_pdf, label="Cauchy Distribution", color="red", linewidth=2)

# Highlight the differences
# plt.ylim(0, 0.5)
plt.axhline(y=0, color='black', linewidth=0.5)
plt.axvline(x=0, color='black', linewidth=0.5)

# Labels and legend
plt.xlabel("x")
plt.ylabel("Probability Density")
# plt.title("Comparison of Normal and Cauchy Distributions")
plt.legend()

# Show the plot
plt.show()
```
